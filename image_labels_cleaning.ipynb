Step 1 : The data is loaded and sanity is checked

```python
import pandas as pd
```

```python
data = pd.read_csv('dummy_image_labels.csv')
print(data.shape)
print(data.dtypes)
```

```python
data.head(5)
```

```python
print(data.isna().sum())
print(data.columns)
```

```python
print (data.iloc[0,1])
```

The label column has very long string which is a list of multiple labels and their respective confidence index
To separate the, this column is split into mutiple columns

```python
new = data["labels (L)"].str.split(",", n = 14, expand = True) 
```

```python
new.head(10)
print(new.shape)
```

Concatenating two dataframes to get one datagrame with its respective image name and the associated labels

```python
separated_data = pd.concat([data, new], axis=1)
```

```python
print(separated_data.shape)
separated_data.head(5)
```

```python
separated_data = separated_data.drop(['labels (L)'], axis=1)
```

```python
print(separated_data.shape)
separated_data.head(5)
```

Gathering all the labels and creating a new dataframe which has an image name and one label at a time

```python
image_label_data = pd.melt(separated_data, id_vars="image_names (S)")
```

```python
print(image_label_data.shape)
image_label_data.head(5)
```

```python
image_label_data = image_label_data.drop(['variable'], axis=1)
print(image_label_data.shape)
image_label_data.head(5)
```

SPlitting value column to seperate label and confidence

```python
image_label_data[['stupid_tag','label_confidence']] = image_label_data.value.str.split(":",expand=True) 
image_label_data.head(5)
```

```python
image_label_data = image_label_data.drop(['value','stupid_tag'], axis=1)
image_label_data.head(5)
```

Final splitting to separate label and confidence

```python
image_label_data[['label','confidence']] = image_label_data['label_confidence'].str.split("-",n = 2,expand=True)
```

```python
image_label_data.head(5)
```

```python
image_label_data.isna().sum()
```

```python
image_label_data = image_label_data.dropna()
image_label_data.isna().sum()
```

Deleting numbers after decimal point from confidence column as extreme precision is not required

```python
image_label_data['confidence'] = [x[:3] for x in image_label_data['confidence']]
```

```python
image_label_data.head(10)
```

Removing punctuation from the label column using following function

```python
import string 
def remove_punctuations(text):
    if text is not None:
        for punctuation in string.punctuation:
            text = text.replace(punctuation, '')
        return text

image_label_data["label"] = image_label_data['label'].apply(remove_punctuations)
```

```python
image_label_data.head(10)
```

```python
image_label_data = image_label_data.drop(['label_confidence'], axis=1)
```

```python
image_label_data.shape
image_label_data.head(5)
```

Checking total labels and counting their occurences.

Saving this data on local machine

```python
labels = image_label_data.groupby(['label']).count()
labels.head(5)
```

```python
labels = labels.drop(['confidence'], axis=1)
labels.rename(columns = {"image_names (S)": "occurrences"}, inplace = True)
labels.head(5)
```

```python
labels.dtypes
```

```python
labels = labels.sort_values(by='occurrences', ascending=False)
labels.head(5)
```

```python
labels.to_excel('label occurrences.xlsx')
```

Saving image name with one label and its confidence, at a time

```python
image_label_data.to_excel('labels_gathered.xlsx')
```

Confidence percentage data which is in object type, is converted into int type in the following code

```python
image_label_data['confidence'] = image_label_data['confidence'].astype(str).astype(int)
```

```python
Spreaded_labels = pd.pivot_table(image_label_data,index='image_names (S)',columns='label',values='confidence')
Spreaded_labels=Spreaded_labels.reset_index()
```
